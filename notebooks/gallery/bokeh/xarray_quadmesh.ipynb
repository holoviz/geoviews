```python
import xarray as xr
import holoviews as hv
import geoviews as gv

hv.extension('bokeh')
```

## Define data

```python
tiles = gv.WMTS('https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{Z}/{Y}/{X}.jpg')
rasm = xr.tutorial.load_dataset('rasm')
qmeshes = gv.Dataset(rasm.Tair[::4, ::3, ::3]).to(gv.QuadMesh, groupby='time')
```

## Plot

```python
%%opts QuadMesh [width=700 height=350]
tiles * qmeshes
```
